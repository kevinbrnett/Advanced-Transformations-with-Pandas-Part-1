```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json
```

```python
# Load data
df = pd.read_csv('C:\\Users\\hamma\\Documents\\DataEnrichment\\Lessons\\Advanced-Transformations-with-Pandas-Part-1\\final_results_crab_cakes.csv.gz')
df.head()
```

## Separate a String Column into Multiple Columns

```python
# Explore existing column format
df['display_phone'].head(2)
```

```python
# Split string and separate into two columns
df['display_phone'].str.split(expand=True)
```

```python
# Save new columns to df
df[['area_code','phone_number']] = df['display_phone'].str.split(' ', expand=True)
df.head(2)
```

```python
# Drop unnecessary columns
df = df.drop(columns=['display_phone','phone'])
df.head(2)
```

## Replacing Multiple Characters at Once within a String Column

```python
# Make list of characters to replace
to_replace = ['(',')']

# Run a loop to replace all the characters in the list at once
for char in to_replace:
    df['area_code'] = df['area_code'].str.replace(char,'',regex=False)

df['area_code'].head(2)
```

## Converting a String Column of Dictionaries into Actual Dictionaries

```python
# Examine single value from the coordinates column
coord = df.loc[0,'coordinates']
print(type(coord))
coord
```

```python
# Replace single quotes with double quotes
coord = coord.replace("'", '"')

# Use json to load coord variable to convert strings to dictionaries
fixed_coord = json.loads(coord)
print(type(fixed_coord))
fixed_coord
```

```python
# Apply to entire column
df['coordinates'] = df['coordinates'].str.replace("'", '"')
df['coordinates'] = df['coordinates'].apply(json.loads)
df['coordinates'].head(3)
```

## Unpack a Column of Dictionaries into Separate Columns

```python
# Convert 'coordinates' column into two columns
lat_long = df['coordinates'].apply(pd.Series)
lat_long
```

```python
# Concatenate to df
df = pd.concat((df, lat_long), axis=1)
df.head(3)
```

```python
# Drop unecessary columns
df = df.drop(columns=['coordinates'])
```

## Convert a Column that Contains Lists (that have been converted to strings) back to lists

```python
# Replace single quotes with double quotes
df['transactions_split'] = df['transactions'].str.replace("'", '"')
```

```python
# Apply json.loads to entire column
df['transactions_split'] = df['transactions_split'].apply(json.loads)

# Check results
df['transactions_split'].head()
```

## Convert a Column of lists into One Hot Encoded Columns

```python
# Exploding column lists
exploded = df.explode('transactions_split')
exploded[['name','transactions','transactions_split']].head()
```

```python
# Saving unique values from exploded column
cols_to_make = exploded['transactions_split'].dropna().unique()
cols_to_make
```

```python
# For loop to create a new column for each unique value
for col in cols_to_make:
    df[col] = df['transactions'].str.contains(col)
df.head()
```

```python
# Drop transaction columns
df = df.drop(columns=['transactions','transactions_split'])

# Save as .csv
df.to_csv('advanced_tf_data_p1.csv', index=False)
```
